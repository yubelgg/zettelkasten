```python
import numpy as np
import time
import matplotlib.pyplot as plt

N = 5   
D = 10
X = np.random.rand(N,D)
print(X)
print(X.shape)

# first function to fill, compute distance matrix using loops
def compute_distance_naive(X):
    N = X.shape[0]      # num of rows
    D = X[0].shape[0]   # num of cols
    print("num of rows: ", N)
    print("num of cols: ", D)
    
    M = np.zeros([N,N])
    for i in range(N):
        for j in range(N):
            xi = X[i,:]
            xj = X[j,:]  
            # expanded formula: np.sqrt(np.sum(xi**2) - 2*np.dot(xi, xj) + np.sum(xj**2))
            # computed matrix are not identical, numerical precision issue?
            # simplified formula: np.sqrt(np.sum((xi - xj)**2))
            dist = np.sqrt(np.sum((xi - xj)**2))  
            M[i,j] = dist

    return M

M = compute_distance_naive(X)
print(M)
```

```python
import numpy as np
import time
import matplotlib.pyplot as plt

N = 5   
D = 10
X = np.random.rand(N,D)


# third function to fill, compute correlation matrix using loops
def compute_correlation_naive1(X):
    N = X.shape[0]  # num of rows
    D = X[0].shape[0]  # num of cols

    # use X to create M
    M = np.zeros([D, D])

    # mean of each column with loop
    mu = np.zeros(D)
    for j in range(D):
        mu[j] = (1/N) * np.sum(X[:, j])
    # print("this is mu: ", mu, sep='\n')

    # covariance matrix with loops
    S = np.zeros([D, D])
    for i in range(D):
        for j in range(D):
            xi = X[:, i]
            xj = X[:, j]
            covariance = (1 / (N - 1)) * np.sum((xi - mu[i]) * (xj - mu[j]))
            S[i, j] = covariance
    # print("this is covariance: ", S, sep='\n')

    # standard deviation
    sigma = np.zeros(D)
    for i in range(D):
        sigma[i] = np.sqrt(S[i, i])  
    print("this is sigma: ", sigma, sep='\n')

    # correlation matrix with loop  
    for i in range(D):
        for j in range(D):
            if sigma[i] * sigma[j] != 0:
                corr = S[i, j] / (sigma[i] * sigma[j])
            else:
                corr = 0
            M[i, j] = corr

    return M

k = compute_correlation_naive1(X)
# print("this is k: ", k, sep='\n')

# fourth function to fill, compute correlation matrix without loops
def compute_correlation_smart(X):
    N = X.shape[0]  # num of rows
    D = X[0].shape[0]  # num of cols

    # use X to create M
    M = np.zeros([D, D])

    # mean of each column
    mu = np.mean(X, 0)
    # print("this is mu: ", mu, sep='\n')

    # covariance matrix without loops
    covariance = np.dot((X - mu).T, (X - mu)) / (N - 1.0)
    # print("this is covariance: ", covariance, sep='\n')

    # standard deviation
    sigma = np.sqrt(covariance.diagonal())
    print("this is sigma: ", sigma, sep='\n')

    # correlation matrix without loops
    sigma_outer = np.outer(sigma, sigma)
    M = covariance / sigma_outer

    return M
M = compute_correlation_smart(X)
# print("this is M: ", M, sep='\n')

assert np.allclose(M, k, atol=1e-06)
```
